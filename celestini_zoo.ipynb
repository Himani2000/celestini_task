```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('zoo.data', sep=",", header=None)
```

```python
data.columns=["animal_name","hair","feathers","egg","milk","airborne","aquatic","predator","toothed","backbone","breathes","venomous","fins","legs","tail","domestic","catsize","type"]
```

```python
data
```

```python
# the dataframe consists of x and y so  splitting the data and the target to x and y  
x=data.iloc[:,1:17]
y=data.iloc[:,17]
```

```python
# splitting the data to train and test split using 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train, y_train)  
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, x_train, y_train, cv=5)
scores            
```

```python
clf.score(x_test, y_test)
```

```python
y_pred=clf.predict(x_test)
```

```python
#### method 2 by using the kfold cross validation 
clf2 = svm.SVC()
```

```python
X=data.iloc[:,1:17]
Y=data.iloc[:,17]
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
from sklearn.model_selection import KFold
cv = KFold(n_splits=5)
scores2=[]
for train_index, test_index in cv.split(X):
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)

    X_train, X_test, Y_train, Y_test = X[train_index], X[test_index], Y[train_index],Y[test_index]
    clf2.fit(X_train, Y_train)
    scores2.append(clf2.score(X_test, Y_test))
```
